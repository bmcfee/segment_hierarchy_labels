```python
import jams
```

```python
def summarize(ann):
    
    h_int, h_lab = jams.eval.hierarchy_flatten(ann)
    
    n_levels = len(h_int)
    
    n_segs = [len(_) for _ in h_int]
    n_labs = [len(set(_)) for _ in h_lab]
    
    return n_levels, n_segs, n_labs
```

```python
jam = jams.load('../data/estimations/SALAMI/410.jams', validate=False)
```

```python
summarize(jam.annotations[0])
```

```python
summarize(jam.annotations[1])
```
